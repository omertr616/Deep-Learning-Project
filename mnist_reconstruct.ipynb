```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import transforms
from torchvision.datasets import MNIST
import torch.nn as nn
import data_utils

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
plt.rcParams.update({'font.size': 12})
device
```

```python
DATA_DIR = '/datasets/cv_datasets/data'
ds_train = MNIST(root=DATA_DIR, download=True, train=True, transform=None)
ds_test = MNIST(root=DATA_DIR, download=True, train=False, transform=None)
dl_train, dl_val, dl_test = data_utils.get_data_loaders(train_dataset=ds_train, test_dataset=ds_test, batch_size=64)
```

```python
from mnist import encoder, decoder

encoder = encoder.encoder_mnist().to(device)
decoder = decoder.decoder_mnist().to(device)
autoencoder = nn.Sequential(encoder, decoder).to(device)

loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(list(encoder.parameters()) + list(decoder.parameters()), lr=5e-4)
```

```python
import training

try:
    encoder.load_state_dict(torch.load('saved_models/encoder_reconstruct_mnist.pth'))
    decoder.load_state_dict(torch.load('saved_models/decoder_reconstruct_mnist.pth'))
except:
    encoder_decoder_trainer = training.AutoencoderTrainer(autoencoder=autoencoder,
                                                            optimizer=optimizer,
                                                            loss_fn=loss_fn,
                                                            device=device,
                                                            transform=None)
            
    encoder_decoder_trainer.train(num_epochs=15,
                                dl_train=dl_train,
                                dl_val=dl_val)
    encoder_decoder_trainer.test(dl_test=dl_test)

    torch.save(encoder.state_dict(), 'saved_models/encoder_reconstruct_mnist.pth')
    torch.save(decoder.state_dict(), 'saved_models/decoder_reconstruct_mnist.pth')
```

```python
from utils import plot_tsne

plot_tsne(encoder, dl_test, device=device, name='reconstruction_mnist')
```

```python
from mnist import classifier

classifier = classifier.ClassifierMnist()
model = nn.Sequential(encoder, classifier).to(device)
```

```python
for param in list(encoder.parameters()):
    param.requires_grad = False
for param in list(classifier.parameters()):
    param.requires_grad = True
```

```python
try:
    model.load_state_dict(torch.load('saved_models/classifier_reconstruct_mnist.pth'))
except:
    optimizer = torch.optim.Adam(model.parameters(), lr=5e-4)
    loss_fn = nn.CrossEntropyLoss()
    trainer = training.ClassifierTrainer(model=model,
                                        optimizer=optimizer,
                                        loss_fn=loss_fn,
                                        device=device)
    trainer.train(num_epochs=20, dl_train=dl_train, dl_val=dl_val)
    trainer.test(dl_test)
    
    torch.save(model.state_dict(), 'saved_models/classifier_reconstruct_mnist.pth')
```
